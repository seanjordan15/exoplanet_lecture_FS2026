```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the data in as a pandas dataframe
df = pd.read_csv('ps_data.csv',header=0,sep=',')
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df['pl_orbper'], df['pl_bmassj'], color='grey', alpha=0.2)

plt.title("Exoplanets Mass vs Period", fontsize=24)
plt.ylabel(r"Mass / M$_{\rm Jupiter}$", fontsize=16)
plt.xlabel("Period / days", fontsize = 16)
plt.yscale("log")
plt.xscale("log")
```

```python
plt.figure(figsize=(10, 6))

# Plot data

plt.title("Exoplanets Mass vs Separation", fontsize=24)
plt.ylabel(r"Mass / M$_{\rm Jupiter}$", fontsize=16)
plt.xlabel("Separation / AU", fontsize=16)
plt.yscale("log")
plt.xscale("log")
plt.legend(fontsize=16, frameon=False, loc='lower right')
```

```python
plt.figure(figsize=(10, 6))

ages = df.st_age.to_numpy()

# check for NaN values
ages = ages[~np.isnan(ages)]

bins = np.linspace((min(ages)),(max(ages)),50)
plt.hist(ages, bins=bins)

# Plot Imaged planets only

plt.title("Exoplanets Age distribution", fontsize=24)
plt.ylabel("Frequency", fontsize=16)
plt.xlabel("Age / Gyrs", fontsize=16)
plt.legend(fontsize=16, frameon=False, loc='upper right')
```
